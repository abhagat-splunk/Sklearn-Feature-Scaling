### Min-Max Scaler
One of the most famous scaling algos.

##### (xi–min(x))/(max(x)–min(x))

Shrinks the range st the range is now between 0 and 1 (or -1 to 1, if there are negative values)
If distribution is not Gaussian or SD is very small, this algo works better.

It is sensitive to Outliers - if many outliers - use RobustScaler instead.

```python
import pandas as pd
import numpy as np
from sklearn import preprocessing
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
df = pd.DataFrame({
    'x1': np.random.chisquare(8,1000),
    'x2': np.random.beta(8,2,1000) * 40,
    'x3': np.random.normal(50,3,1000)
})

scaler = preprocessing.MinMaxScaler()
scaled_df = scaler.fit_transform(df)
scaled_df = pd.DataFrame(scaled_df, columns=['x1','x2','x3'])

fig, (ax1, ax2) = plt.subplots(ncols=2, figsize=(6, 5))
ax1.set_title('Before Scaling')
sns.kdeplot(df['x1'], ax=ax1)
sns.kdeplot(df['x2'], ax=ax1)
sns.kdeplot(df['x3'], ax=ax1)
ax2.set_title('After Min-Max Scaling')
sns.kdeplot(scaled_df['x1'], ax=ax2)
sns.kdeplot(scaled_df['x2'], ax=ax2)
sns.kdeplot(scaled_df['x3'], ax=ax2)
plt.show()
```
